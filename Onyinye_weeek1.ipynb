# probability that a randomly selected registered female will be enrolled in the program

```python
no_female_registered = (reg_df['GENDER'] == "Female").sum()
total_registerd = len(reg_df["GENDER"])
#no_female_registered
#total_registerd
#prob_reg_female_enr = 
no_female_enrolled = (enr_df['GENDER'] == "Female").sum()
prob_enr_female = (no_female_enrolled/no_female_registered)*100
prob_enr_female
```

### The probabilty that a randomly selected registered female will be enrolled is 19.94 percent

# Probabilty that a randomly selected male will be enrolled in the program

```python
no_male_registered = (reg_df['GENDER'] == "Male").sum()

no_male_enrolled = (enr_df['GENDER'] == "Male").sum()
prob_enr_male = (no_male_enrolled/no_male_registered)*100
prob_enr_male
```

### The probabilty that a randomly selected registered male will be enrolled is 55.91 percent

# Probability that a registerd applicant will eventually enroll in the program

```python
total_registerd = len(reg_df["GENDER"])
total_enrolled = len(enr_df["GENDER"])

prob_of_enrollement = (total_enrolled/total_registerd)*100
prob_of_enrollement
```
